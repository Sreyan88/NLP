```python
def text_process(mess):
    
    nopunc = [char for char in mess if char not in string.punctuation]

    nopunc = ''.join(nopunc)
    
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
messages['message'].head(5).apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=text_process).fit(messages['message'])
print(len(bow_transformer.vocabulary_))
```

```python
message4 = messages['message'][3]
print(message4)
```

```python
bow4 = bow_transformer.transform([message4])
print(bow4)
print(bow4.shape)
```

```python
label1=pd.factorize(messages['label'])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
start_time = time.time()
classifier = Sequential()
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 50, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 50, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11425))
classifier.add(Dropout(0.2))

classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(messages_tfidf, label1[0], batch_size = 10, epochs = 10)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
y_pred = classifier.predict(messages_tfidf)
```
